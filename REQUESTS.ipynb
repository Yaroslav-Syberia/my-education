```python
def currency_name(currency_id):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    currencies = requests.get(url).json()['Valute']
    for currency in currencies.values():
        if currency['ID'] == currency_id:
            return currency['Name']
```

```python
def get_smm_index(group_name, token):
    # Получаем количество участников группы
    url = 'https://api.vk.com/method/groups.getMembers'  
    params = {  
        'group_id': group_name,  
        'v': 5.95,  
        'access_token': token  
    }  
    response = requests.get(url, params = params).json()  
    group_count = response['response']['count'] # количество участников группы
    # Получаем последние 10 сообщений на стене группы
    url = 'https://api.vk.com/method/wall.get'  
    params = {  
        'domain': group_name,  
        'filter': 'owner',  
        'count': 10,  
        'offset': 0,  
        'access_token': token,  
        'v': 5.95  
    }  
    response = requests.get(url, params = params).json()
    # Сохраняем статистики последних сообщений в словарь
    stats = {}
    for record in response['response']['items'][:]:  
        title = record['text'][:30]  
        if title:  
            stats[title] = [record['comments']['count'], record['likes']['count'], record['reposts']['count'], record['date'] ]  
    # Считаем smm_index как сумму лайков, комментов, репостов и делим на количество участников группы
    smm_index = sum([sum(post[:3]) for post in stats.values()]) / group_count
    return smm_index

get_smm_index('mountelbrus', token)
```
